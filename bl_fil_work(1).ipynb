```python
import sigpyproc
import h5py
import numpy as np
import matplotlib.pyplot as plt
from sigpyproc.readers import FilReader
```

```python
fil_file=FilReader('breakthrought_listen/bl_voyager/voyager_f1032192_t300_v2.fil')
```

```python
fil_file.header
```

```python
#Processing the data with sigpyproc by 1st reading the data block
data = fil_file.read_block(0,2 )
tdownsamp=1
subband=512
#Next downsample the data as there are 6,00,000 channels
data = data.downsample(tdownsamp, subband)
```

```python
plt.figure(figsize=(40,20))
plt.subplot(2,1,1)
plt.xlabel('Channels')
plt.ylabel('Arbitrary units')
plt.title('Voyager telemetry data is one of the spikes (I have not converted channels to MHZ)')
plt.plot(data[:,1])

plt.subplot(2,1,2)
plt.xlabel('Channels')
plt.ylabel('Arbitrary units')
plt.title('Voyager telemetry data is one of the spikes (I have not converted channels to MHZ)')
plt.plot(data[:2*10**4,1])
```
