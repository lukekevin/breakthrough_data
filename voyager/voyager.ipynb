```python
import blimpy as bl
#import matplotlib.pyplot as plt
import pylab as plt
import numpy as np
import scipy as sp
%matplotlib inline
```

```python
fil_file=bl.Waterfall('bl_filterbank/v_fiterbank.fil')
```

```python
fil_file.info()
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(4,1,1)
fil_file.plot_spectrum(f_start=8420.193, f_stop=8420.195) # left sideband
plt.subplot(4,1,2)
fil_file.plot_spectrum(f_start=8420.2163, f_stop=8420.2166) # carrier
plt.subplot(4,1,3)
fil_file.plot_spectrum(f_start=8420.238, f_stop=8420.24) # right sideband
plt.subplot(4,1,4)
fil_file.plot_spectrum(f_start=8420.18, f_stop=8420.25)
plt.tight_layout()
plt.savefig('v_tel_subplots.png')
```
